### 영화 평점 데이터 분석
- 사용자별 영화 평점 구하기

```python
#데이터 읽어오기
data= np.genfromtxt('data/ratings.txt',delimiter="::",dtype='int')
data
#사용자번호,영화번호,영화평점,영화시간
```

```python
#배열의 크기확인
data.shape
#행,열
```

```python
#전체 영화 평점 평균(2열) => data.mean()
data[:,2].mean()
#data.sum()=>합계
```

```python
#사용자별 평균평점
#사용자 아이디가 1인 데이터
user1=data[data[:,0]==1]
user1[:,2].mean()
```

- 전체 사용자의 평점평균 각 구해보기

```python
#사용자 수 확인(중복제거)
len(data[:,0])
user_uni=np.unique(data[:,0])
len(user_uni)
```

```python
for i in user_uni:
    user_id=data[data[:,0]==i]
    user_mean=user_id[:,2].mean()
    print(f"{i}번 사용자 평점: {user_mean}")
    
    # print(f"{i}번 사용자 평점: {user[:,2].mean()}")
    # print(f"{i}번 사용자 평점: {data[data[:,0]==i][:,2].mean()}")
```
