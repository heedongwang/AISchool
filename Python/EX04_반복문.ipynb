```python
import random as rd
ran= rd.randint(1,50)

while True:
    num=int(input("숫자를 입력하세요>>"))
    if ran>num:
        print(f"{num}보다 큰 수 입니다.")
    elif ran<num:
        print(f"{num}보다 작은 수 입니다.")
    else:
        print("정답을 맞추셨습니다.")
        break
```
