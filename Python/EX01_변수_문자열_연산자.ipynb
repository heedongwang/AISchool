### 문자열 인덱싱 슬라이싱
- 문자열도 인덱스 번호를 가진다
- 인덱싱: 인덱스 번호 하난를 가리켜서 데이터 추출
- 슬라이싱: 범위를 지정하여 여러개의 데이터 추출

```python
name = 'My name is YH'
```

```python
#n 추출하기
# 인덱싱 방법: 변수명[인덱스 번호]=> 인덱스 번호는 0부터 시작, 띄어쓰기도 인덱스 번호 가짐
print(name[3])
print(name[11])
# 뒤에서부터 인덱스번호도 가능(-1,-2...)
name[-2]
```

```python
#슬라이싱 방법: 변수명[범위]=> [시작인덱스: 끝인덱스+1]
# 범위 작성시 끝인덱스 범위는 가지고오고싶은 인덱스 +1
print(name[0:2])
print(name[:2])

print(name[3:7])

print(name[11:13])
print(name[11:])
print(name[-2:])
```

# 문자열 포매팅
1. %기호 포매팅: % 기호를 이용하여 숫자를 대입=> 단점: 데이터 타입에 따른 알파벳 암기
2. format 함수: {} .format() =>중괄호 순서와 포맷함수 안에 들어가는 변수 순서가 일치해야함
3. f문자열 포매팅:{변수}, 직관적이다


```python
month=6
day=25
today='오늘은 month월 day일 입니다'
today
# 문자열 연산을 할때 문자열과 숫자는 연산이 안된다.=>포매팅사용
```

```python
# 1.%기호
month=6
day=25
today='오늘은 %d월 %d일 입니다'%(month,day)
today
```

```python
#2.format
# 중괄호 순서와 포맷함수 안에 들어가는 변수 순서가 일치해야함
month=6
day=25
today='오늘은 {}월 {}일 입니다'.format(month,day)
today
```

```python
# 3.f문자열 포맷팅
    
month=6
day=25
today=f'오늘은 {month}월 {day}일 입니다'
today
```

```python
x= 100
y=200
add=x+y
print(f'{x}와 {y}의 합은 {add}입니다')
```
