### 리스트 함수
1. 리스트 값 추가= append
2. 원하는 인덱스 위치에 값을 추가=insert
3. 리스트에 있는 값 수정=> 리스트명[수정하고 싶은 인덱스]=새로운값
4. 리스트 내 값 삭제= del, remove
5. 정렬= sort
6. 리스트 내 요소 개수 파악=>len()

```python
#1.append
list2=[0,1,2,3,4,5]
#리스트 가장 뒤에 값을 하나씩 추가하는 함수.
#리스트명.append(value)
list2.append(6)
list2

#리스트 안에 다른 리스트 삽입가능
list2.append([0,1,2])
list2
```

```python
f_list = ['사과','포도',['수박','멜론'],'복숭아','딸기','오렌지']

choice_list=[]

apple=f_list[0]
waterMelon=f_list[2][0]
orange=f_list[-1]

choice_list.append(apple)
choice_list.append(waterMelon)
choice_list.append(orange)

choice_list

```

```python
#2.insert
#리스트명.insert(인덱스 번호, 값), 자동으로 인덱스 밀림
#choice에서 사과와 수박 사이에 포도 넣기
choice_list.insert(1,'포도')
choice_list
```

```python
#music_list = [['Hype boy','NewJeans'],['LOVE DIVE','아이브'],['사건의 지평선','윤하']]
music_list
loc= int(input("인덱스 위치 입력>>"))
song= input("노래 제목 입력>>")
singer= input("가수 입력>>")

music_list.insert(loc,[song,singer])

```

```python
#3.리스트 값 수정
#리스트명[수정하고 싶은 인덱스]=새로운값
list2[0]=7
list2
```

```python
#4.리스트 삭제
#del
del list2[3]
```

```python
#remove
#리스트명.remove(값)
list2.remove(6)
```

```python
#5.정렬
#sort() 기본 오름차순(reverse=False), 내림차순(reverse=True)
list3=[2,45,12,59,48,85]

list3.sort() #오름차순
list3.sort(reverse=True)
list3
```

```python
#리스트 내 요소 개수 파악
#len(리스트명)
len(list3)
```
