```python
#라이브러리 불러오기
import pandas as pd
```

```python
#Series=>pandas 자료구조 중 하나로 인덱스와 값이 1:1로 대응하는 1차원 자료구조
population = pd.Series([9668465,3391946,2942828,1450062],
                      index = ['서울','부산','인천','광주'])
population
```

# Series 자료구조 확인

```python
#series값만 확인 => 넘파이
population.values
```

```python
#series 인덱스만 확인
population.index
```

```python
#Series 이름 지정
population.name='인구'
population
```

```python
#인덱스에 이름 지정
population.index.name="도시"
population
```
