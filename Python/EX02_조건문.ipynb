```python
score=int(input("점수 입력>>"))

if 90<=score<=100:
    grade="A"
elif 80<=score:
    grade="B"
elif 70<=score:
    grade="C"
elif 60<=score:
    grade="D"
else:
    grade="F"
print(f"{score}점은 {grade}학점입니다.")
```

```python
print('''
================================================
================---- 자판기 ----================
================================================
== 1.콜라 == 2.사이다 ==  3.물  =======insert===
==--------==----------==--------=======------===
==   600  ==    800   ==  1000  ======= O00원===
=======================================------===
================================================
==++++++++++++++++++++++++++++++++++++++++++++==
==++++++++++++++++++++++++++++++++++++++++++++==
==++++++++++++++++++++++++++++++++++++++++++++==
==+++++++++++++++              +++++++++++++++==
==+++++++++++++++      AIS     +++++++++++++++==
==+++++++++++++++              +++++++++++++++==
==++++++++++++++++++++++++++++++++++++++++++++==
==++++++++++++++++++++++++++++++++++++++++++++==
================================================
================================================
  ===                                      ===
  ===                                      ===
''')

# 돈입력
money = int(input("insert coin>>"))
# 메뉴선택
choice = input("메뉴선택")
# 메뉴선택 후 잔액계산
if choice == '1':
    change = money - 600
elif choice == '2':
    change =  money - 800
elif choice == '3':
    change = money - 1000
else:
    print("메뉴를 잘 못 선택하셨습니다~")

# 거스름돈 계산 -> 출력
if change < 0 :
    print("잔액이 부족합니다")
else:
    ch = change // 1000
    o = change%1000//500
    b = change%1000%500//100
    print(f"천원{ch}개 오백원{o}개 백원{b}개")
```
