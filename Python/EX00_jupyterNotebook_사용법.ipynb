### 주피터 노트북 사용법
#### 마크다운 모드
- 셀하나를 통째로 코드영향을 받지 않는 메모장 형태로 만들때 사용
- 커멘드모드에서 단축기 M ->실행하면 메모장 형태로 저장
- 


### 셀 상태 변경
- command mode: 셀 단위 작업시 사용
- edit mode: 코드 작성시 사용
- command-> eidt:  마우스로 코드부분 클릭 or 엔터
- eidt-> command: 셀 윗부분 클릭, esc

### 단축기 정리(command mode에서 셀단위 작업)
- b: 현재 셀 아래 셀 추가 
- a: 현재 셀 위에 셀 추가
- dd: 셀 삭제
- m: 마크다운
- y: 마크다운 모드에서 코드 셀로 변경
